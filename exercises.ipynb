## Second exercise: top 10 arrival airports in the world in 2013 

```python
arrival_airports= bookings.groupby('arr_port', sort=True)['pax'].agg(['sum']).rename(columns={'sum':'nb_passengers'})
top10_airports=arrival_airports.sort_values('nb_passengers',ascending=False).head(10)
top10_airports
```

## Third exercise: plot the monthly number of searches for flights arriving at MÃ¡laga, Madrid or Barcelona

```python
searches.Date= pd.to_datetime(searches.Date,yearfirst=True)
searches['month'] = pd.DatetimeIndex(searches['Date']).month
```

```python
dest_agp_bcn_mad= searches[(searches['Destination']=="AGP'") | (searches['Destination']=="MAD") | (searches['Destination']=="BCN")]
dest_agp_bcn_mad.head(10)
```

```python
dest_agp_bcn_mad.set_index('month', inplace=True)
p=dest_agp_bcn_mad.groupby('Destination').count().plot(legend=True)
p.figure.show()
```
